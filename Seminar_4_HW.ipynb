### 1.1. Данные из прошлого дз

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv', sep = ',', encoding='latin-1')
df.head()
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(6, 6))
plt.hist(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6, 8))
sns.histplot(df['sqft_living'], bins=20)
plt.title('Распределение квадратуры жилой')
plt.xlabel('Кв. метры жилые')
plt.ylabel('Количество')
```

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(6, 8))
sns.histplot(df['yr_built'], bins=30)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество')
```

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.scatter(df['price'], df['sqft_living'])
plt.title('Взаимосвязь между ценой и жилой площадью')
plt.xlabel('Цена(миллионы)')
plt.ylabel('Жилая площадь')
plt.show()
```

*Большинство домов находятся в пределах 2.5 миллионов и до 5000 квадратных метров. Так же можем отметить выбросы, например один из домов за 2.5 миллиона имеет площадь почти в 14.000 квадратных метров, или же например один из домов с 12.000 квадратными метрами имеет цену в 7.7 миллионов. Тем не менее, в общем наблюдается достаточно устойчивая кореляция между площадью и ценой. **Чем больше площадь,тем выше цена***

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

*Чем больше жилая площадь, тем выше  стоимость дома*

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

*Чем выше качество конструкции и дизайн дома, тем выше его стоимость, в основном дома сосредоточены в 7 классе и ценой до миллиона*

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Распределение цен по количеству этажей');
```

*Наибольшую стоимость имеют дома с количеством этажей 2,5*

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Распределение цен с оценкой вида');
```

*Наибольшую стоимость имеют дома с оценкой вида 4*
